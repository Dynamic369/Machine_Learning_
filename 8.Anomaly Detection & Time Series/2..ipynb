## DBSCAN Clustring for outliers

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
X ,y = make_circles(n_samples=750, factor=0.3, noise=0.1)
# factor determines the relative size of the inner circle compared to the outer circle.
```

```python
plt.scatter(X[:,0], X[:,1],c =y)
```

```python
dbscan = DBSCAN(eps=0.125)
```

```python
prediction = dbscan.fit_predict(X)
```

```python
prediction
```

```python
import numpy as np
index = np.where(prediction < 0)
```
