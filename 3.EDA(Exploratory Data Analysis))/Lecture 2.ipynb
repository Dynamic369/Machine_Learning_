```python
# importing the important libraries
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('Data_Train.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date']=df['Date_of_Journey'].str.split("/").str[0]
df['Months']=df['Date_of_Journey'].str.split("/").str[1]
df['Year']=df['Date_of_Journey'].str.split("/").str[2]
```

```python
df.drop('Date_of_Journey',inplace=True,axis = 1)
```

```python
# converting the str datatype into integer datatype
df['Date'] = df['Date'].astype(int)
df['Months'] = df['Months'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.head()
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
df.drop('Dep_Time',axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Duration_hour'] = df['Duration'].str.split(" ").str[0].str.split("h").str[0].astype(int())
df['Duration_min'] = df['Duration'].str.split(" ").str[1].str.split('h').str[0].astype(int())
```

```python
# Checking the presence of any string in the particular column.
df['Duration_hour'].str.contains(r'.+').unique()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(":").str[0].astype(int)
df['Arrival_min'] = df['Arrival_Time'].str.split(":").str[1].str.split(' ').str[0].astype(int)
```

```python
df.drop("Arrival_Time",axis = 1, inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stop'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3 ,'4 stops':4, np.nan:1})
```

```python
df.drop("Total_Stops", axis=1, inplace = True)
```
